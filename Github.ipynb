```python
def collect_repositories(start_date, end_date, keyword='data science', per_page=100):
    repositories = []
    page = 1
    time.sleep(1)
    while True:
        try:
            url = f'https://api.github.com/search/repositories?q={keyword} created:{start_date}..{end_date}&per_page={per_page}&page={page}' 
            r = requests.get(url)
            json_pydata = r.json() # Shortcut: json_pydata = r.json()

            for item in json_pydata['items']:
                repositories.append(item)

            print(f"Num remaining requests: {r.headers['X-RateLimit-Remaining']}")
            page += 1
        except Exception as e: # Có thể là do hết lượng request được phép, hoặc hết page, ...
            print(f'Error: {e}')
            break
    return repositories 
```

```python
start_date = '2018-01-01'
end_date = '2018-01-15'
keyword = 'data science'
per_page = 100
repositories = collect_repositories(start_date, end_date, keyword, per_page)
print(len(repositories))
```

```python
for data in repositories:
    data.pop('owner', None)
    data.pop('license', None)
```

```python
df = pd.DataFrame(repositories)
df.to_csv('repositories.csv', encoding='utf-8', index=False)
```
